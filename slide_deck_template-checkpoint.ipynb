```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('ford_gobike.csv')
```

## Distribution of Genders

> About three-quarters of Ford GoBike's riders are males by 74.6%, then females are 23%, and other genders are 2.1%. 

```python
# sort values based on member_gender 
sorted_counts = df['member_gender'].value_counts()

# figure size
plt.figure(figsize = [6,4])

# set pie chart
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False,autopct='%1.1f%%');
plt.rcParams['font.size'] = 12
plt.title("The percentage of people who ride Ford GoBike's bisycles based on person's gender \n", fontsize=12)
plt.axis('square');
```

## Distribution of Trip Durations

>  The most frequent trip durations is between 4-6 minutes, and after 6 minutes of the trip duration, the frequency decreases as we increasing the trip duration.

```python
# figure size
plt.figure(figsize = [6, 4])

# set histogram chart
bins=np.arange(0,df['duration_min'].max()+1,1)
plt.hist(data=df,x='duration_min',bins=bins);
plt.xlim(1,40)
plt.title('The frequency of trip duration in minutes \n' , fontsize=12)
plt.xlabel('Duration in minutes')
plt.ylabel('Frequency');
```

## Distribution of Ages
> From the age of 20 up to 75, most of the people who ride Ford GoBike's bicycles are between the age of 26 and 35, and the frequency decreases after the age of 33.

```python
# figure size
plt.figure(figsize = [22, 10])

# set Bar chart
bars_color = sb.color_palette()[0]
bars_order = df['member_age'].value_counts().index
sb.countplot(data=df,x='member_age',color=bars_color);
plt.title("The frequency of Ford GoBike's bicycles riders based on user's age \n (between age of 20 and 75) \n ")
plt.xlabel('User Age')
plt.ylabel('Count')
plt.xlim(0,55);
```

## Trip Duration and Rider Gender
> Females are spending more trip duration than Males by a few different, the median of Females trip duration is 10 minutes and colse to 30 minutes as a maximum, and for Males the median is about 9 minutes, and the maximum reaches 24 minutes, without taking outliers in consideration. (Excluding other genders)

```python
# figure size
plt.figure(figsize = [6,4])

# set Box chart
base_color = sb.color_palette()[0]
ax=sb.boxplot(data=df, x='member_gender', y='duration_min', color=base_color);
plt.title("The comparison between rider's gender and the trip duration \n",fontsize=12)
plt.xlabel('Gender ')
plt.ylabel('Duration in Minutes')
plt.ylim(0,60)
ax.set_yticks([0,5,10,15,20,25,30,35,40,45,50,55,60]);
```

## Trip Durations by Rider Gender and Age Stage
> In all age stages between 20s and up to 70s, females spend more trip duration than males except male who's age stage in the 70s, he spend lesser time than females in the same age stage. And most of the trip duration of these ages stages are between 10 minutes up to 15 minutes. (Excluding other genders)

```python
# figure size
plt.figure(figsize = [6, 4])

# set clustered bar chart
ax = sb.barplot(data = df, x = 'age_stage', y = 'duration_min', hue = 'member_gender')
ax.legend(loc = 8, ncol = 3, framealpha = 1, title = 'Member Gender');
plt.title("Trip duration compared to rider's age stage with rider's gender \n",fontsize=12)
plt.xlabel('Age Stage')
plt.ylabel('Duration in minutes');
```
